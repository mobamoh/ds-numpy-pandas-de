# Visualisierung in pandas

Es gibt zahlreiche Bibliotheken in Python zum Erstellen von Visualisierungen verfügbar. Oftmals werden wir wahrscheinlich [Matplotlib](http://matplotlib.org/) und/oder [Seaborn](http://stanford.edu/~mwaskom/software/seaborn/) für alles verwenden, was allgemeiner Natur ist, und dann andere Bibliotheken, wenn wir etwas Spezielleres benötigen ([Plotly](https://plot.ly/) für Dashboards, zum Beispiel). All diese Bibliotheken ermöglichen es uns, großartig aussehende Visualisierungen zu erstellen, die in einer Produktionsumgebung verwendet werden können. Wenn wir etwas Schnelles und Grobes wollen, um unsere Daten sehr einfach zu visualisieren, gibt es auch einige Plotting-Funktionalitäten, die in Pandas eingebaut sind.

Wenn wir uns die [Dokumentation](https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html) ansehen, können wir sehen, dass das in pandas verfügbare Plotting über die `plot()`-Methode auf einem DataFrame-Objekt aufgerufen wird. Von dort aus werden wir eine Reihe potenzieller Argumente an die `plot()`-Methode übergeben, um genau anzugeben, wie der Plot erstellt werden soll. Das wichtigste dieser Argumente ist das `kind`-Schlüsselwortargument, das der `plot()`-Methode mitteilt, welche Art von Visualisierung wir möchten (Balkendiagramm, Histogramm, Streudiagramm usw.). Die meiste Zeit werden wir deine Visualisierung in Matplotlib oder Seaborn machen, aber hier ist ein kleiner Vorgeschmack darauf, was pandas tun kann, damit du weißt, dass es da ist.

Zuerst müssen wir pandas und unseren DataFrame importieren.

```python
import pandas as pd
df = pd.read_csv('data/winequality-red.csv', delimiter=';')
```

```python
df.head()
```

```python
df.columns
```

Wir haben mehrere Eingabevariablen; sie basieren auf physikalisch-chemischen Tests:
- 1 - fixed acidity (feste Säure)
- 2 - volatile acidity (flüchtige Säure)
- 3 - citric acid (Zitronensäure)
- 4 - residual sugar (Restzucker)
- 5 - chlorides (Chloride)
- 6 - free sulfur dioxide (freies Schwefeldioxid)
- 7 - total sulfur dioxide (Gesamtschwefeldioxid)
- 8 - density (Dichte)
- 9 - pH
- 10 - sulphates (Sulfate)
- 11 - alcohol (Alkohol)

Und es gibt eine Ausgabevariable, basierend auf sensorischen Daten:
- 12 - quality (Qualität, eine Bewertung zwischen 0 und 10)

Um einen guten Überblick über die Daten zu bekommen, wollen wir einige Histogramme zeichnen. Wir können dies mit dem Argument `kind = 'hist'` visualisieren. Wir müssen auch angeben, welche Dichteverteilung gezeichnet werden soll, also müssen wir eine bestimmte Spalte benennen. Die Ausgabevariable, die Qualität des Weins, scheint am interessantesten zu sein, also zeichnen wir sie zuerst.

```python
# specify a certain column and call the pandas plot function
df['quality'].plot(kind='hist')
```

Du kannst sehen, dass die Rotweine in diesem Datensatz Qualitätsbewertungen von 3 bis 8 haben. Wir haben also weder Weine von sehr schlechter Qualität noch Weine mit den besten Qualitätsbewertungen. Die Bewertung 5 wird am häufigsten vergeben. Fast 700 Weine haben diese Bewertung erhalten.

Auch andere Plots können mit der Plot-Funktion erstellt werden. Als Nächstes versuchen wir Streudiagramme, bei denen wir ein Feature für die X- und Y-Spalte angeben müssen. Und setzen `kind='scatter'`.

```python
# As you might guess from the error, we have to specify X and Y columns for pandas to plot. 
df.plot(kind='scatter', x='total sulfur dioxide', y='free sulfur dioxide')
```

So sehen Streudiagramme für kontinuierliche Variablen aus. Du kannst sehen, dass wenn das Gesamtschwefeldioxid zunimmt, auch das freie Schwefeldioxid zunimmt.
Es ist auch interessant zu sehen, wie verschiedene Features die Qualität von Rotweinen beeinflussen. Qualität ist eine diskrete Variable, also sei nicht überrascht, dass der Plot anders aussieht.

```python
df.plot(kind='scatter', x='quality', y='alcohol')
```

Dies ist möglicherweise nicht der beste Plot, um viele Schlussfolgerungen zu ziehen, aber zumindest sehen wir, dass Weine mit höherer Qualität keine niedrigen Alkoholkonzentrationen haben.

Die nächste Art von Plot, die wir uns ansehen wollen, sind Boxplots. Wie Histogramme sollten sie uns einen besseren Eindruck von der Verteilung der Daten geben.

```python
df.plot(kind='box')
```

Für die meisten Features können wir ihre Verteilung nicht unterscheiden, dies liegt an den unterschiedlichen Skalen der Features. Daher ist es notwendig, nur einige Spalten anzugeben und zu sehen, wie das aussieht...

```python
df[['fixed acidity', 'pH', 'alcohol']].plot(kind='box')
```

```python
# This still doesn't look great - it's hard to really examine these three columns since pH is 
# so different from the other two. Let's drop pH and try one more time...
df[['fixed acidity', 'alcohol']].plot(kind='box')
```
