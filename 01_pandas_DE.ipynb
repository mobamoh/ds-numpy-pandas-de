## Ein DataFrame-Objekt erhalten

Es gibt zwei grundlegende Möglichkeiten, wie wir ein pandas DataFrame-Objekt zum Arbeiten bekommen können. Die erste besteht darin, Daten zu verwenden, die bereits in unserem Python-Programm vorhanden sind, in Verbindung mit dem `DataFrame`-Konstruktor. Die zweite besteht darin, externe Daten über das pandas-Modul einzulesen (das wir importiert und über `pd` zugänglich gemacht haben). Zur Referenz, hier sind die [Dokumentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html) für pandas DataFrames.

##### Verwendung von Daten, die bereits in unserem Python-Programm vorhanden sind

Wenn wir Daten verwenden, die bereits in unserem Python-Programm vorhanden sind, werden wir diese Daten an den `DataFrame`-Konstruktor übergeben. Wir machen dies typischerweise auf zwei Arten. Die erste beinhaltet das Übergeben einer Liste von Dictionaries, während die zweite das Übergeben von zwei Listen beinhaltet. Lass uns mit der ersten beginnen...

````python
import pandas as pd # We haven't actually done this in code yet. 
data_lst = [{'a': 1, 'b': 2, 'c':3}, {'a': 4, 'b':5, 'c':6, 'd':7}]
df = pd.DataFrame(data_lst) 
df
````

DataFrames bestehen aus **Zeilen** und **Spalten**, wobei die Spalten durch ihre Namen gekennzeichnet und aufgerufen werden, und die Zeilen durch die Indizes des DataFrame gekennzeichnet und aufgerufen werden. Oben können wir unsere Spalten betrachten und sehen, dass ihre Namen `a`, `b`, `c` und `d` sind. Wir können ähnlich unsere Zeilen betrachten und sehen, dass sie mit `0` und `1` indiziert sind. Diese Spaltennamen und Indizes sind die Art und Weise, wie wir später auf diese Daten zugreifen werden. Wie hat der `DataFrame`-Konstruktor unsere Liste von Dictionaries genommen und sie in diesem Format in den DataFrame eingefügt?

Wenn der pandas DataFrame-Konstruktor auf eine Liste von Dictionaries wie die, die wir ihm gegeben haben, stößt, interpretiert er jedes Dictionary als eine Zeile im DataFrame. Die Schlüssel werden als Spaltennamen gelesen und die Werte als die Werte für jede Spalte. Standardmäßig weist der DataFrame-Konstruktor eine Spalte für **jeden** Schlüssel zu, den er in **irgendeinem** Dictionary in der Liste von Dictionaries sieht. Wenn ein bestimmtes Dictionary in dieser Liste keinen Wert für diesen Schlüssel hat, dann weist es einen `NaN`-Wert (steht für "not a number") für dieses Index-Spalten-Paar zu. Als der pandas DataFrame oben also die Liste von Dictionaries bekam, sah er die Schlüssel `a`, `b`, `c` und `d` und erstellte daher diese Spalten. Dann füllte er die Werte ein, die mit diesen Schlüsseln verbunden waren, und füllte ein `NaN` ein, wenn er diesen Schlüssel nicht fand (wie er `d` nicht im ersten Dictionary in unserer Liste fand).

````python
data_lst = [{'a': 1}, {'b':5}, {'c': 4}]
df = pd.DataFrame(data_lst)
````

**Was erwartest du, dass unser DataFrame jetzt enthält?**

````python
df
````

Die zweite Möglichkeit, einen DataFrame aus Daten zu erstellen, die bereits in unserem Python-Programm vorhanden sind, besteht darin, eine Liste von Listen als `data`-Argument und eine Liste von Strings als `columns`-Argument zu übergeben. Der `pd.DataFrame()`-Konstruktor nimmt an, dass jede einzelne Liste im `data`-Argument eine Zeile ist (d.h. wenn du eine Liste von 5 Listen übergibst, wird dein DataFrame 5 Zeilen haben). Unten übergeben wir eine Liste von 2 Listen an den `data`-Parameter, was bedeutet, dass unser DataFrame zwei Zeilen haben wird.

````python
data_vals = [[1, 2, 3], [4, 5, 6]]
data_cols = ['a', 'b', 'c']
df = pd.DataFrame(data=data_vals, columns=data_cols)
df
````

Es ist wichtig zu beachten, dass diese Methode nicht ganz so flexibel ist wie die Verwendung einer Liste von Dictionaries. Beim Übergeben einer Liste von Listen über das `data`-Argument müssen wir sicherstellen, dass die größte Anzahl von Elementen in einer einzelnen Liste der Anzahl der Spaltennamen entspricht, die wir über das `columns`-Argument übergeben (nicht mehr, nicht weniger). Zum Beispiel:

````python
data_vals = [[1, 2], [4, 5, 6]]
data_cols = ['a', 'b']
df = pd.DataFrame(data=data_vals, columns=data_cols)
````

Wir haben jedoch die Flexibilität, einige Zeilen mit 'fehlenden' Daten zu übergeben. Der wichtige Punkt hierbei ist, dass der letzte Spaltenname derjenige wird, der mit einem `NaN` gefüllt wird, wenn einer Liste ein Spaltenwert fehlt (es basiert nicht auf einem Schlüsselnamen wie in unserer Liste von Dictionaries).

````python
data_vals = [[1, 2], [4, 5, 6]]
data_cols = ['a', 'b', 'c']
df = pd.DataFrame(data=data_vals, columns=data_cols)
df
````

#### Einlesen externer Daten

Es gibt viele Möglichkeiten, wie wir externe Daten in einen pandas DataFrame einlesen können, und sie werden als Funktion aufgerufen, die über das `pandas`-Modul verfügbar ist.

Als Ergebnis des Imports des `pandas`-Moduls und seiner Zugänglichmachung über `pd` bedeutet dies, dass wir alle diese Funktionen über `pd` aufrufen werden. Jede dieser Funktionen wird uns ein pandas DataFrame-Objekt **zurückgeben**, gefüllt mit den externen Daten, die wir eingelesen haben.

Die [pandas-Dokumentation](https://pandas.pydata.org/pandas-docs/stable/user_guide/io.html) zeigt dir alle Möglichkeiten, wie du externe Daten in einen DataFrame laden könntest. Also, wenn wir Daten aus einem CSV laden wollten, würden wir einfach verwenden:

```python

```

Hinweis: Dies setzt voraus, dass wir die Spaltennamen in der ersten Zeile deiner .csv haben.

Wenn wir die Spaltennamen nicht in der ersten Zeile unserer .csv haben, könnten wir die .csv mit folgendem einlesen:

Wenn wir die Spaltennamen beim Einlesen zuweisen wollten, können wir ein zusätzliches `names`-Argument übergeben, wobei dieses `names`-Argument eine Liste der Namen enthält, die wir den Spalten zuweisen möchten.

```python
df = pd.read_csv('my_data.csv', header=None, names=['col1', 'col2', ...., 'col12'])
```  

Das standardmäßig angenommene Spaltentrennzeichen ist ein Komma. Falls deine CSV-Datei ein anderes verwendet, kannst du den `sep`-Parameter verwenden (oder `delimiter`, welches ein Alias von `sep` ist). Es ist eine notwendige Fähigkeit zu erkennen, dass es möglicherweise bereits eine Lösung für solche Probleme gibt. Tatsächlich ist `sep` in der [Dokumentation](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html) beschrieben.

```python
df = pd.read_csv('my_data.csv', ..., sep=';')
```

#### Die Grundlagen

Wir wissen jetzt, wie wir unsere Daten betrachten können. Was ist, wenn wir bestimmte Teile zum Betrachten oder bestimmte Teile zum Herumspielen/Transformieren greifen wollten? Sagen wir, wir wollten eine ganze Zeile oder eine ganze Spalte greifen... wie machen wir das? Lass uns eintauchen, indem wir mit etwas Indizierung beginnen.

Das Format, das wir verwenden, um in unseren DataFrame zu indizieren und Daten zu greifen, hängt davon ab, welche Teilmenge der Daten wir genau greifen wollen. Wenn wir ganze Zeilen oder Spalten greifen wollen, dann können wir dafür Klammernotation verwenden (genau wie wir Klammernotation verwenden, um in Listen zu indizieren). Wenn wir eine ganze Spalte wollen, dann werden wir den **Spaltennamen** in Klammern setzen (und mehrere Spaltennamen in einer Liste innerhalb dieser Klammern). Wir können manchmal auch über Punktnotation auf eine Spalte auf dem DataFrame zugreifen, was wir gleich zeigen werden. Wenn wir eine ganze Zeile wollen, dann müssen wir **sowohl** einen **Anfangs- als auch Endindex** innerhalb der Klammern platzieren (es funktioniert nicht, nur einen einzelnen Index in die Klammern zu setzen).

````python
# Grabs the "sex" column: bracket notation
df['sex']
````

````python
# Grabs the "sex" column: dot notation
df.sex
````

>Hinweis: Die Punktnotation funktioniert nicht, wenn die Spaltennamen Leerzeichen enthalten! Deshalb ist es praktisch, zuerst zu prüfen, ob Spaltennamen Leerzeichen enthalten und diese in "_" (Unterstriche) zu ändern.

Glücklicherweise haben die Spalten in unserem DataFrame bereits Unterstriche.
Falls nicht, hier ist ein kurzes Code-Snippet, das du verwenden könntest, um die Spaltennamen zu ändern:

````python
# make a copy of your DataFrame
df2 = df.copy()
# make a list column names
cols = df2.columns.tolist()
# replace space with _
cols = [col.replace(' ', '_') for col in cols]
# reassign new column names to DataFrame
df2.columns = cols
````

Was ist, wenn wir bestimmte Zeilen **und** bestimmte Spalten greifen wollen, anstatt nur ganze Zeilen oder ganze Spalten?

Wenn wir nur bestimmte Zeilen und Spalten greifen wollen, gibt es zwei **Methoden**, die wir verwenden können, um in einen pandas DataFrame zu indizieren: `loc[]` und `iloc[]`. Beachte, dass dies **Methoden** sind, was bedeutet, dass wir sie über Punktnotation auf unserem `DataFrame`-Objekt aufrufen werden. Der Unterschied zwischen diesen dreien hat damit zu tun, wie wir sie verwenden. `loc[]` ist ein rein label-positionsbasierter Indexer, `iloc[]` ist ein rein integer-positionsbasierter Indexer.

````python
# Loc is label based. All of these will work, because they are recognized as labels on the 
# rows (index labels) or columns (column name labels). 
df.loc[0, 'sex'] # 0 is one of the index labels, and 'sex' is a column label.
````

````python
# Ranges on our index labels still work (as long as they're numeric).
df.loc[0:10, 'sex']
````

````python
df.loc[10:15, ['sex', 'length']]
````

````python
# These will all fail, because they attempt to access the columns by position integers, 
# and loc only takes labels. 
df.loc[0, 0]
df.loc[0:10, 0]
df.loc[10:15, [0, 4]]
````

````python
# The above will all work with .iloc, though, since it takes integers (and not labels)
df.iloc[0, 0]
df.iloc[0:10, 0]
df.iloc[10:15, [0, 4]]
````

````python
# Using labels, though, like we did with .loc, will NOT work.
df.iloc[0, 'sex']
````

#### Ein bisschen mehr

Wir wissen, wie man bestimmte Zeilen oder Spalten aus DataFrames greift, sowie eine Teilmenge von Zeilen und Spalten und alles dazwischen. Aber das sieht so aus, als würde es typischerweise erfordern, dass die genaue Position der Daten, die wir wollen, bekannt ist. Was ist, wenn wir diese Position nicht kennen? Gibt es eine Möglichkeit, gewünschte Daten zu greifen, indem wir einfach einige Abfrageparameter spezifizieren? Ja!

Es gibt ein paar Möglichkeiten, wie wir dies tun können. Die erste Möglichkeit, die wir uns ansehen werden, ist einfach durch Maskierung, während die zweite tatsächlich die `query()`-Methode verwendet, die auf dem pandas DataFrame verfügbar ist.

````python
# This just gives us a mask - tells us True or False whether each row fit's the condition.
df['length'] <= 0.4
````

````python
# To use a mask, we actually have to use it to index into the DataFrame (using square brackets). 
df[df['length'] <= 0.4]
````

Beachte, wie nur die Indizes, die bei der Bedingung als True befunden wurden, in dieser Teilmenge des DataFrame erscheinen. Wir haben den Rest der Indizes "maskiert", die als False befunden wurden (daher der Name **Maskierung**).

````python
# Okay, this is cool. What if I wanted a slightly more complicated query...
df[(df['length'] <= 0.4) & (df['sex'] == 'F')]
````

````python
# So I could write an arbitrarily complicated query using that syntax... 
df[(df['length'] <= 0.4) & (df['length'] > 0.35) & (df['weight_whole'] > 0.25) & (df['weight_whole'] < 0.3)]
````

````python
# Or I could use the query() method that is available on our DataFrame object. 
df.query('length <= 0.4 and length > 0.35 and weight_whole > 0.25 and weight_whole < 0.3')
````

Im Allgemeinen wird es bevorzugt, die `query()`-Methode zu verwenden, da sie die Lesbarkeit verbessert. Sie verwendet nicht viele Sätze von eckigen Klammern (`[]`) und Klammern (`()`), sondern nur einen Satz Klammern. Sie folgt auch tendenziell der Python-Syntax etwas näher als die Maskierungsmethoden, die wir oben betrachtet haben, indem sie `and` anstelle von `&` verwendet, um verschiedene Spezifikationen in unseren Abfragen zu trennen.

#### Groupby

Beginnen wir mit `groupby`s...

Mit "group by" beziehen wir uns auf einen Prozess, der einen oder mehrere der folgenden Schritte beinhaltet:

- **Aufteilen** der Daten in Gruppen basierend auf einigen Kriterien
- **Anwenden** einer Funktion auf jede Gruppe unabhängig
- **Kombinieren** der Ergebnisse in eine Datenstruktur

[Quelle](https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html)

````python
# first we try to groupby the sex, let's check how many different values this feature can have 
df['sex'].unique()
````

````python
# we can also see the count of different values with this command
df['sex'].nunique()
````

````python
df.groupby('sex')   # Note that this returns back to us a groupby object. It doesn't actually 
                    # return to us anything useful until we perform some aggregation on it. 
````

Wir haben tonnenweise Aggregationsmetriken, die wir von einem groupby-Objekt bekommen können. Beachte hier, dass wir die Ergebnisse eines groupby unten speichern, um dann alle Arten von Operationen darauf durchzuführen (dies ist tatsächlich die bevorzugte Methode, wenn wir mehr als eine Berechnung darauf durchführen werden).

````python
# Group by sex column
groupby_obj = df.groupby('sex')

# Performing an aggregation by calculating the mean
groupby_obj.mean()
````

````python
groupby_obj.max()
````

````python
groupby_obj.count()
````

Die vorherigen Aggregationsmetriken gaben uns einen DataFrame mit allen Spalten minus der, nach der wir gruppiert haben, zurück. Beachte, dass das, wonach wir gruppiert haben, zum Index wird. Was ist, wenn ich nur eine Spalte zurück haben wollte (besonders bei etwas wie count, wo es für jede Spalte gleich ist)?
Nun, wir können mit diesem DataFrame alles machen, was wir vorher gemacht haben...

````python
df.groupby('sex').count()['length']
````

Beachte, dass wir auch nach mehreren Spalten gruppieren können, indem wir sie in einer Liste übergeben. Es wird zuerst nach der ersten übergebenen Spalte gruppiert und dann nach der zweiten danach (d.h. es wird nach der zweiten innerhalb der Gruppierung der ersten gruppiert).

````python
# Note we can also group by multiple columns by passing them in a list. It will group by 
# the first column passed in first, and then the second after that (i.e. it will group by 
# the second within the group by of the first). 
df.groupby(['sex', 'n_rings']).count()['weight_whole']
````

Schaue dir die [Group By Dokumentation](https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html) an, um zu sehen, was du alles mit dem pandas `.groupby()` machen kannst.

````python
df.sort_values('length').head(15) # Note: this is ascending by default.
# df.sort_values('length', ascending=False) # here we change ascending to false.
````

Wir können nach mehreren Spalten sortieren, indem wir sie in eine Liste innerhalb der sort()-Methode setzen. Es wird zuerst nach der ersten übergebenen Spalte sortiert und dann nach der zweiten innerhalb der Sortierung der ersten.

#### Spalten erstellen und löschen

Spalten werden auf eine von zwei Arten erstellt:
1. Verwendung von Klammernotation
2. Verwendung der `eval()`-Methode auf dem pandas DataFrame.

Das Löschen von Spalten erfolgt mit der `df.drop()`-Methode auf dem pandas DataFrame. Du musst auch sicherstellen, der `drop()`-Methode mitzuteilen, auf welcher Achse sich das zu löschende Ding befindet (Zeilen sind `axis=0` und Spalten sind `axis=1`).

### Umgang mit Nullwerten

pandas hat Funktionen sowohl zum Füllen von Nullwerten (oder N/As) mit einem beliebigen Wert als auch zum vollständigen Löschen von Nullwerten. Um Nullwerte zu füllen, verwenden wir die `.fillna()`-Methode auf dem DataFrame, und um Nullwerte zu löschen, rufen wir die `.dropna()`-Methode auf dem DataFrame auf. In Bezug auf die `.fillna()`-Funktion können wir ihr einen Standardwert zum Auffüllen geben oder eine Reihe anderer Methoden zum Auffüllen (Padding, Rückwärtsfüllen, Vorwärtsfüllen). Du kannst über den Umgang mit fehlenden Daten in der Dokumentation [hier](https://pandas.pydata.org/pandas-docs/stable/user_guide/missing_data.html) lesen. Wir werden hier nicht zu sehr in die Tiefe gehen, aber wollen, dass du weißt, dass diese Funktionalität existiert.

Wie wir bereits überprüft haben, gibt es in diesem Datensatz nicht einmal fehlende Daten. Die nächste Codezeile wird in diesem Fall also nichts tun. Nur damit du weißt, wie man NANs mit pandas füllt oder löscht.

````python
# Notice the addition of the inplace argument here. 
# if you try to fill nans will some value: the first argument is the inserted value
df.fillna(-1, inplace=True)
# dropna() will drop the whole row, if nans are present
df.dropna(inplace=True) 
````
