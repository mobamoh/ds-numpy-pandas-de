```python
import pandas as pd
from IPython.display import display

red = pd.read_csv("data/winequality-red.csv", sep=";")
white = pd.read_csv("data/winequality-white.csv", sep=";")
```

3. Ich habe beschlossen, dass ich diesen Monat Weine mit relativ hohem Alkoholgehalt meiden möchte. Um das zu tun, werde ich alle Weine meiden, die einen höheren Alkoholgehalt als den durchschnittlichen Alkoholgehalt haben, und du wirst mir dabei helfen. Um dies zu erreichen, lass uns Folgendes tun:

  * Finde den durchschnittlichen Alkoholgehalt, getrennt, für Rot- und Weißweine.
  * Erstelle eine `Series`, die angibt, ob jede Zeile in jedem `DataFrame` (die Rot- und Weißweine) einen höheren Alkoholgehalt als den Durchschnitt hat.
  * Merge diese `Series` auf den `DataFrame`. Ich kann mir vorstellen, dies entweder mit einem `.join()` oder mit `pd.concat()` zu tun. Mach es zur Übung mit beiden. Hinweis: Merges mit `Series` funktionieren auf die gleiche Weise wie sie mit `DataFrames` funktionieren.
  * Gib mir alle die Zeilen zurück, die mir helfen, diese Weine mit höherem Alkoholgehalt zu meiden.

```python
red_alcohol_mean = red["alcohol"].mean()
white_alcohol_mean = white["alcohol"].mean()

red_high_alcohol = (red["alcohol"] > red_alcohol_mean).rename("high_alcohol")
white_high_alcohol = (white["alcohol"] > white_alcohol_mean).rename("high_alcohol")

red_joined = red.join(red_high_alcohol)
white_joined = white.join(white_high_alcohol)

red_concat = pd.concat([red, red_high_alcohol], axis=1)
white_concat = pd.concat([white, white_high_alcohol], axis=1)

display(pd.Series({
    "red_alcohol_mean": red_alcohol_mean,
    "white_alcohol_mean": white_alcohol_mean,
}))
display(red_joined[~red_joined["high_alcohol"]].head())
display(white_joined[~white_joined["high_alcohol"]].head())
display(red_concat[~red_concat["high_alcohol"]].shape)
display(white_concat[~white_concat["high_alcohol"]].shape)
```

4. Sagen wir, dass ich nächsten Monat anfangen möchte, mich einzuschränken. Dieses Mal möchte ich mich jedoch darauf konzentrieren, Weine mit hoher Säure zu meiden. Speziell möchte ich diese Weine meiden, die im höchsten Bin der festen Säure sind (höchstes Bin von 5). Du wirst mir jetzt dabei helfen. Um dies zu erreichen, lass uns Folgendes tun:

 * Trenne die Zeilen in jedem `DataFrame` in 5 gleich breite Bins (nicht gleich Quintilen) basierend auf ihrer festen Säure.
 * Merge die resultierende `Series`, die diese 5 Bins enthält, auf den ursprünglichen `DataFrame`. Ich kann mir vorstellen, dies auch entweder mit `.join()` oder mit `pd.concat()` zu tun. Versuche es zur Übung mit beiden zu tun.
 * Gib mir alle die Zeilen zurück, die **nicht im** obersten Bin in Bezug auf feste Säure sind.

```python
red_fixed_acidity_bin = pd.cut(red["fixed acidity"], bins=5).rename("fixed_acidity_bin")
white_fixed_acidity_bin = pd.cut(white["fixed acidity"], bins=5).rename("fixed_acidity_bin")

red_binned_join = red.join(red_fixed_acidity_bin)
white_binned_join = white.join(white_fixed_acidity_bin)

red_binned_concat = pd.concat([red, red_fixed_acidity_bin], axis=1)
white_binned_concat = pd.concat([white, white_fixed_acidity_bin], axis=1)

red_top_bin = red_fixed_acidity_bin.cat.categories[-1]
white_top_bin = white_fixed_acidity_bin.cat.categories[-1]

display(red_binned_join[red_binned_join["fixed_acidity_bin"] != red_top_bin].head())
display(white_binned_join[white_binned_join["fixed_acidity_bin"] != white_top_bin].head())
display(red_binned_concat[red_binned_concat["fixed_acidity_bin"] != red_top_bin].shape)
display(white_binned_concat[white_binned_concat["fixed_acidity_bin"] != white_top_bin].shape)
```

5. Sagen wir, dass ich jetzt wissen möchte, wie sehr meine Entscheidung, diese Weine mit höherem `alcohol`-Gehalt zu meiden, die `quality` der Weine einschränken wird, die ich trinken kann. Um das herauszufinden, möchte ich ein paar Dinge wissen:

 * Der durchschnittliche `alcohol`-Gehalt für die Rotweine über dem durchschnittlichen `alcohol`-Niveau, nach Qualität.
 * Der durchschnittliche `alcohol`-Gehalt für die Weißweine über dem durchschnittlichen `alcohol`-Niveau, nach Qualität.
 
 Verwende eine `pivot table`, um dies zu lösen.

```python
display(pd.pivot_table(
    red_joined[red_joined["high_alcohol"]],
    values="alcohol",
    index="quality",
    aggfunc="mean",
))

display(pd.pivot_table(
    white_joined[white_joined["high_alcohol"]],
    values="alcohol",
    index="quality",
    aggfunc="mean",
))
```

6. Mach jetzt das Gleiche für meine Entscheidung, nächsten Monat Weine mit hoher Säure zu meiden:

 * Finde den durchschnittlichen `alcohol`-Gehalt für Rotweine, nach `quality` und `fixed acidity` Bin.
 * Finde den durchschnittlichen `alcohol`-Gehalt für Weißweine, nach `quality` und `fixed acidity` Bin.
 
  Verwende eine `pivot table`, um dies zu lösen.

```python
display(pd.pivot_table(
    red_binned_join,
    values="alcohol",
    index="quality",
    columns="fixed_acidity_bin",
    aggfunc="mean",
))

display(pd.pivot_table(
    white_binned_join,
    values="alcohol",
    index="quality",
    columns="fixed_acidity_bin",
    aggfunc="mean",
))
```
