## Teil 2: NumPy arrays manipulieren.

1. Bitte erstelle ein NumPy array mit (1, 3, '4', 7, 12, '0'). Definiere den Inhalt als Integer.

`````python
arr = np.array((1, 3, '4', 7, 12, '0'), dtype=int)
arr
`````

2. Überprüfe den Datentyp der Objekte und die Form des Arrays

`````python
arr.dtype, arr.shape
`````

3. Aktualisiere den 4. Wert auf 30.

`````python
arr[3] = 30
arr
`````

4. Forme das Array in eine 2x3-Matrix um.

`````python
matrix = arr.reshape(2, 3)
matrix
`````

5. Bitte addiere 8 zu jedem Wert in der ersten Zeile und 12 zu jedem in der zweiten Zeile.

`````python
matrix = matrix + np.array([[8], [12]])
matrix
`````

6. Multipliziere die erste Spalte mit 2, die zweite mit 3 und die dritte mit 4.

`````python
matrix = matrix * np.array([2, 3, 4])
matrix
`````

7. Bitte summiere alle Zahlen in der ersten Zeile und alle Zahlen in der zweiten Zeile.

`````python
matrix.sum(axis=1)
`````

8. Suche ähnlich die größte Zahl für jede Spalte.

`````python
matrix.max(axis=0)
`````

9. Extrahiere die Zahl in der zweiten Spalte und der ersten Zeile.

`````python
matrix[0, 1]
`````

10. Überprüfe, bei welchem Index der Wert genau 48 ist.

`````python
np.argwhere(matrix == 48)
`````
