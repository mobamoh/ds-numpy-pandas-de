Pandas-Funktionen, die es uns ermöglichen, zwei Datensätze zu kombinieren, umfassen die Verwendung von `pd.merge()`, `df.join()`, `df.merge()` und `pd.concat()`. Zum größten Teil tun diese weitgehend die gleichen Dinge (obwohl du den leichten Syntaxunterschied bei `merge()` und `concat()` bemerken wirst, die über das pandas-Modul aufgerufen werden können, und `merge()` und `join()`, die auf einer DataFrame-Instanz aufgerufen werden können). Es gibt einige Fälle, in denen eines davon besser sein könnte als ein anderes, in Bezug auf das Schreiben von weniger Code oder das Durchführen einer Art von Datenkombination auf eine einfachere Weise. Die Hauptunterschiede zwischen diesen hängen jedoch weitgehend davon ab, was sie standardmäßig tun, wenn du versuchst, verschiedene Daten zu kombinieren. Standardmäßig versucht `merge()`, über gemeinsame Spalten zu verbinden, `join()` über gemeinsame Indizes und `concat()`, indem es einfach auf einer gegebenen Achse anhängt.

Du kannst mehr Details über die Unterschiede zwischen allen drei in der [Dokumentation](https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html) finden. Wir werden uns unten einige Beispiele ansehen.

`````python
# We'll go back to our wine data set. Who doesn't love wine?
import pandas as pd
wine_df = pd.read_csv('data/winequality-red.csv', delimiter=';')
wine_df.head()
`````

`````python
# A glance at the values of the quality of wine in the DataFrame
wine_df.quality.unique()
`````

`````python
# get_dummies is a method called on the pandas module - you simply pass in a pandas Series 
# or DataFrame, and it will convert a categorical variable into dummy/indicator variables. 
quality_dummies = pd.get_dummies(wine_df.quality, prefix='quality')
quality_dummies.head()
`````

Schauen wir uns nun die `join()`-Methode an. Denk daran, diese verbindet standardmäßig über Indizes. Das bedeutet, dass wir unseren quality dummies DataFrame einfach mit folgendem zurück zu unserem ursprünglichen wine DataFrame verbinden können...

`````python
joined_df = wine_df.join(quality_dummies)
joined_df.head() 
`````

Die Argumente von `.join` sind die folgenden:
````
DataFrame.join(self, other, on=None, how='left', lsuffix='', rsuffix='', sort=False)
````
Mit `how` können wir angeben, welche Join-Methode wir verwenden möchten.

Das how-Argument für merge spezifiziert, wie bestimmt wird, welche Schlüssel in die resultierende Tabelle einbezogen werden sollen. Wenn eine Schlüsselkombination weder in der linken noch in der rechten Tabelle erscheint, werden die Werte in der verbundenen Tabelle NA sein. Hier ist eine Zusammenfassung der how-Optionen und ihrer SQL-äquivalenten Namen:

Merge-Methode | SQL Join Name | Beschreibung
---|---|---
left| LEFT OUTER JOIN | Verwende Schlüssel nur vom linken Frame
right | RIGHT OUTER JOIN | Verwende Schlüssel nur vom rechten Frame
outer | FULL OUTER JOIN | Verwende Vereinigung der Schlüssel von beiden Frames
inner | INNER JOIN | Verwende Schnittmenge der Schlüssel von beiden Frames


Du kannst es auch als Mengenlehre betrachten und Venn-Diagramme verwenden, um zu veranschaulichen, was bei jeder Methode passiert.

![Join-Methoden](./images/join_types.png)

Schauen wir uns nun concat an.
Ähnlich wie bei `join` können wir die Methode angeben, die wir verwenden möchten, um die Datensätze zu kombinieren.

Anders als join und merge, die standardmäßig auf Spalten operieren, kann concat definieren, ob auf Spalten oder Zeilen operiert werden soll.
In den Bildern unten kannst du die Unterschiede sehen, wenn axis auf 0 oder 1 gesetzt ist.

**Concat mit axis=0:**
![Concat Axis 0](./images/concat_axis_0.png)

**Concat mit axis=1:**

`````python
joined_df2 = pd.concat([quality_dummies, wine_df], axis=1)
joined_df2.head()
`````

Lass uns einen anderen Datensatz einlesen, da wir uns das Kombinieren mehrerer Datenquellen ansehen.

`````python
red_wines_df = pd.read_csv('data/winequality-red.csv', delimiter=';')
white_wines_df = pd.read_csv('data/winequality-white.csv', delimiter=';')
`````

`````python
red_wines_df.columns
`````

`````python
white_wines_df.columns
`````

`````python
red_wines_quality_df = red_wines_df.groupby('quality').mean()['fixed acidity'].reset_index()
red_wines_quality_df.head()
`````

`````python
white_wines_quality_df = white_wines_df.groupby('quality').mean()['fixed acidity'].reset_index()
white_wines_quality_df.head()
`````

`````python
pd.merge(red_wines_quality_df, white_wines_quality_df, on=['quality'], suffixes=[' red', ' white'])
`````

`````python
pd.concat([red_wines_quality_df, white_wines_quality_df['fixed acidity']], axis='columns', join='inner')
`````

`````python
red_wines_quality_df.join(white_wines_quality_df, lsuffix='red', rsuffix='white')
`````
