## Die Grundlagen des Array

### Was ist das Besondere an NumPy Arrays?

Was ist so besonders an einem `NumPy array`?

Aus einer hohen Perspektive sind sie irgendwie wie Listen - sie speichern nur einen Haufen Zeug in einem Container. Es stellt sich jedoch heraus, dass ein NumPy array viel schneller ist, um damit zu interagieren und bestimmte Arten von Berechnungen durchzuführen als eine Standard-Python-Liste. Warum ist das so? Die beiden Hauptgründe, warum sie schneller sind, sind:

1. Sie werden als ein zusammenhängender Speicherblock gespeichert, anstatt wie eine Liste über mehrere Speicherorte verteilt zu sein.
2. Jedes Element in einem NumPy array ist vom gleichen Datentyp (d.h. alle Integer, alle Floats usw.), anstatt ein Konglomerat aus beliebig vielen Datentypen (wie es eine Liste ist). Wir nennen diese Idee Homogenität, im Gegensatz zur möglichen Heterogenität von Python-Listen.

Wie viel schneller sind sie? Lass uns die Zahlen von 0 bis 1 Million nehmen und diese Zahlen summieren, wobei wir es sowohl mit einer Liste als auch mit einem NumPy array timen.

```python
import numpy as np # Standard alias when importing NumPy - follow this convention!
numpy_array = np.arange(0, 1000000)
python_list = range(1000000)
```

```python
python_list
```

```python
%timeit np.sum(numpy_array)
```

```python
%timeit sum(python_list)
```

```python
%timeit sum(numpy_array)
```

Also läuft NumPy fast eine Größenordnung schneller! Das liegt an diesen beiden oben genannten Punkten. Da NumPy arrays Daten in zusammenhängenden Speicherblöcken speichern, können sie **Vektorisierung** nutzen, was die Fähigkeit einer CPU ist, eine Operation auf mehreren Datenstücken gleichzeitig durchzuführen. Da ein NumPy array außerdem weiß, welchen Typ jedes Objekt hat, das es speichert (und diese Typen sich nicht ändern), muss es keine Zeit damit verschwenden, zu überprüfen, welchen Typ jedes Objekt hat (wie eine Liste). Die Kombination dieser beiden Dinge beschleunigt unsere Berechnung ziemlich.

Beachte auch, dass wir `np.sum()` angeben mussten - NumPys Implementierung von sum. Als wir einfach die eingebaute Python `sum()`-Funktion auf das NumPy array verwendeten, war die Berechnung tatsächlich langsamer! Das liegt daran, dass NumPy arrays für vektorisierte Berechnungen optimiert sind, und wenn wir versuchen, eine nicht-vektorisierte Operation durchzuführen, zahlen wir einen Preis.

Es ist auch erwähnenswert, dass alles, was wir oben getan haben, eine Summe war - nur eine **einfache** Summe. Wenn wir zu komplizierteren Operationen übergehen, sparen wir noch mehr Zeit! Schauen wir uns an, was NumPy arrays sonst noch können...

### Ein NumPy Array erstellen

Jetzt, da wir wissen, wie großartig `NumPy arrays` sein können, lass uns in sie eintauchen. Wir werden nicht alles abdecken, was du mit `NumPy arrays` machen kannst (siehe die [Methoden-Dokumentation](http://docs.scipy.org/doc/numpy-1.10.0/reference/generated/numpy.ndarray.html#numpy.ndarray) dafür), aber wir werden uns die Grundlagen ansehen.

Beginnen wir damit, wie wir ein `NumPy array` erstellen können. Dazu verwenden wir den `np.array()`-Konstruktor, der irgendeine Art von Array oder etwas erwartet, das die Array-Schnittstelle bereitstellt (d.h. sich wie ein Array verhält - Listen und Tupel sind Beispiele). Das bedeutet also, dass wir ein NumPy array erstellen können, indem wir eine Liste oder ein Tupel übergeben.

```python
my_lst_ndarray = np.array([1, 2, 3, 4, 5])
# You can specify the data type upon creation. 
my_tuple_ndarray = np.array((1, 2, 3, 4, 5), np.int32) 
```

Genau wie wir das shape-Attribut auf pandas DataFrames haben, haben wir es auch auf NumPy arrays.

```python
print(my_lst_ndarray.shape)
print(my_tuple_ndarray.shape)
```

Wir haben auch das dtype-Attribut, das uns den Datentyp der Objekte in unserem ndarray (n-dimensionales Array) mitteilt.

```python
print(my_lst_ndarray.dtype)
print(my_tuple_ndarray.dtype)
```

```python
my_lst_ndarray2 = np.array(["1", 2, 3, "10", 5])
print(my_lst_ndarray2.dtype)
```

"U" steht für Unicode-String und "2", weil es ein 2 Zeichen langer String ist. Jedes Element im Array wurde in einen String konvertiert.

```python
# every element is as type a string now
type(my_lst_ndarray2[1])
```

Wenn du versuchst, dem ndarray mitzuteilen, ein bestimmter Datentyp zu sein, wird es versuchen, jedes Objekt, das du übergibst, in diesen Datentyp zu casten (hier ein 32-Bit-Integer), und scheitern, wenn es ihn nicht in diesen Datentyp casten kann. Unten können wir "10" in einen 32-Bit-Integer casten, also ist das in Ordnung.

```python
my_lst_ndarray3 = np.array([1, 2, 3, "10", 5], np.int32) 
print(my_lst_ndarray3.dtype)
```

```python
# This will not work, because Python can't cast the string 'bozo' as a 32 bit integer. 
my_lst_ndarray3 = np.array([1, 2, 3, "bozo", 5], np.int32) 
```

## NumPy Array Mathematik

Beim Arbeiten mit einem `NumPy` array haben wir alle grundlegenden mathematischen Operatoren zur Verfügung: `+`, `-`, `*`, `/`, `**` und `%`. Häufig werden wir mit zwei Arrays arbeiten, die die gleiche Größe haben, in welchem Fall diese Operatoren einfach **elementweise** durchgeführt werden. Zum Beispiel:

```python
first_arr = np.array([1, 2, 3, 4])
second_arr = np.array([5, 6, 7, 8])
first_arr + second_arr # Each element is lined up with it's corresponding element in the other 
                       # array, and the addition is then performed. 
```

```python
first_arr = np.array([[1, 2], [3, 4]]) # This is now a two-dimensional array. 
second_arr = np.array([[5, 6], [7, 8]]) # This is now a two-dimensional array. 
first_arr * second_arr # Each element is lined up with it's corresponding element in the other 
                       # array, and the multiplication is then performed. 
```

Es stellt sich heraus, dass unsere numerischen Operationen auch funktionieren können, wenn wir eine Operation zwischen einem `NumPy array` und einem einzelnen Wert durchführen wollen. Zum Beispiel, sagen wir, dass wir `4` von `first_arr` oben subtrahieren wollen, oder es mit `5` multiplizieren wollen, oder den Rest finden wollen, wenn alles durch `3` geteilt wird. Wir können das über das Folgende tun:

```python
first_arr = np.array([[1, 2], [3, 4]]) 
```

```python
first_arr - 4
```

```python
first_arr * 5
```

```python
first_arr % 3
```

Das Konzept, das dies ermöglicht, wird als [Broadcasting](https://numpy.org/doc/stable/user/basics.broadcasting.html) bezeichnet. Es ist ein Konzept, das besonders nützlich beim Arbeiten und Interagieren mit `NumPy` arrays sein wird. Grundsätzlich nimmt es diese einzelne Zahl auf der rechten Seite (die `4`, `5` oder `3` oben) und **broadcastet** ihre Form, um mit der von `first_arr` (`2 x 2`) übereinzustimmen. Danach führt es dann die Operation elementweise durch, wie wir es vorher gesehen haben.

Es stellt sich heraus, dass die Dinge etwas komplizierter werden können als dies. Wenn wir wollten, könnten wir mathematische Operationen wie die obigen auf Spalten- oder Zeilenebene durchführen. Zum Beispiel könnten wir `4` von der ersten Spalte und `5` von der zweiten Spalte subtrahieren, oder `4` von der ersten Zeile und `5` von der zweiten Zeile. Wir würden das über das Folgende tun:

```python
first_arr = np.array([[1, 2], [3, 4]])
```

```python
# Here, we subtract 4 off the first column and 5 off the second column. 
first_arr - [4, 5]
```

```python
# Here, we subtract 4 from the first row and 5 from the second row. 
first_arr - [[4], [5]]
```

## Ein bisschen mehr über NumPy Arrays

Okay, also jetzt, da wir ein bisschen über ein `NumPy array` wissen, was können wir sonst noch damit machen? Es gibt tatsächlich ziemlich viele Dinge, die wir tun können. Wir können in sie indizieren, Berechnungen durchführen, nach Aggregations-Typ-Metriken fragen usw.

### Indizierung

Beginnen wir damit, in sie zu indizieren. Bei `NumPy arrays` haben wir nicht die `.loc[]`- oder `.iloc[]`-Methoden wie bei einem DataFrame - wir indizieren einfach in sie wie in eine Liste. Es ist effektiv eine mehrdimensionale Liste. Daher können wir ihr mehrere Indizierungswerte übergeben. Schauen wir uns das an.

```python
# Reshape will reshape the data to the shape that you tell it to (here 5 rows, 4 columns). 
range_arr = np.arange(0, 20, 1).reshape(5, 4)
range_arr
```

```python
# Grab every row, but only the element at index 2 in those rows. 
range_arr[:, 2]
```

```python
# With no second index, this defaults to taking the rows. 
range_arr[0:2]
```

```python
# The first set of numbers refers to the rows to grab, the second set the columns.
range_arr[0:2, 1:3] 
```

### Andere Methoden

Schauen wir uns nun einige der anderen verfügbaren Methoden an. Wieder gibt es eine Menge, was wir tun können, und wir zielen hier darauf ab, dir zumindest einen Überblick über viele der möglichen Dinge zu geben. Wir möchten dir auch ein Notebook hier geben, auf das du zurückblicken kannst, um zu sehen, was möglich ist (Google ist auch dafür erstaunlich).

Wir können Summen in jede Richtung mit einer Methode auf den Arrays durchführen.

```python
# Sum along the rows (i.e. get column totals)
range_arr.sum(axis=0)
```

```python
# Sum along the columns (i.e. get row totals)
range_arr.sum(axis=1)
```

```python
# Get sum of all elements in NumPy array. 
range_arr.sum() 
```

Wir können auch die Mittelwert-, Standardabweichungs-, Max- und Min-Werte entlang der Zeilen (d.h. für die Spalten) greifen. Wir könnten dies auch entlang der Spalten oder für das Array als Ganzes tun (genau wie wir es mit `.sum()` gemacht haben).

```python
range_arr.mean(axis=0)
```

```python
range_arr.std(axis=0)
```

```python
range_arr.max(axis=0)
```

```python
range_arr.min(axis=0)
```

Wenn wir stattdessen den **Index** greifen wollen, an dem diese Min- und Max-Werte auftreten (entweder entlang der Zeilen oder Spalten), dann können wir die `argmin()`- und `argmax()`-Methoden verwenden, die auf unserem NumPy array verfügbar sind.

```python
# We see that the mins of each column occur at row 1 (index 0).
range_arr.argmin(axis=0) 
```

```python
# We see that the maxs of each column occur at row 5 (index 4).
range_arr.argmax(axis=0) 
```

```python
 # Here we get the index of the overall minimum (the 0th index).
range_arr.argmin()
```

```python
# Here we get the index of the overall maximum (the last index). 
range_arr.argmax() 
```

Wir können die kumulative Summe oder das Produkt mit folgendem erhalten.

```python
# Here it gets the cumsum along the rows (i.e. from top to bottom)
range_arr.cumsum(axis=0)
```

```python
# Gets the cumprod along the rows
range_arr.cumprod(axis=0) 
```

```python
# We can flatten our arrays as follows. 
range_arr.flatten()
range_arr.ravel()  # They look the same in this case. 
```

## Ein kurzer Blick auf eine coole NumPy-Methode

Die Mehrheit, wenn nicht alle Methoden, die wir für NumPy arrays betrachtet haben, sind auf pandas-Spalten verfügbar. Sie haben möglicherweise etwas andere Benennungskonventionen (`idxmax` auf einer Spalte versus `argmax` auf einem NumPy array, zum Beispiel), aber da pandas DataFrames auf NumPy arrays aufgebaut sind, überschneiden sich die auf NumPy arrays verfügbaren Methoden weitgehend mit den auf pandas DataFrames verfügbaren Methoden.

Viele dieser Methoden sind auch als Funktionen auf dem `NumPy`-Modul selbst verfügbar. Genau wie wir die `argmax()`-Methode auf einem NumPy array aufrufen können, können wir `np.argmax()` aufrufen und eine Liste oder ein Tupel übergeben. Bevor wir zu DataFrames zurückkehren, schauen wir uns eine letzte Methode an, die in `NumPy` verfügbar ist, `np.where()`. `np.where()` kann uns helfen herauszufinden, welche Elemente in einem NumPy array eine bestimmte Bedingung erfüllen.

```python
my_ndarray = np.array([2, 4, 6, 8, 24, 3, 8, 9, 12])
```

```python
# Returns the indices where the data meet the condition.
print(np.where(my_ndarray <= 2))  
print(np.where(my_ndarray == 8)) 
print(np.where(my_ndarray > 6)) 
```

## Pivot-Tabellen

Aus [wiki](https://de.wikipedia.org/wiki/Pivot-Tabelle): "Neben anderen Funktionen kann eine Pivot-Tabelle automatisch sortieren, Gesamtsummen zählen oder den Durchschnitt der in einer Tabelle oder Tabellenkalkulation gespeicherten Daten geben und die Ergebnisse in einer zweiten Tabelle anzeigen, die die zusammengefassten Daten zeigt. Pivot-Tabellen sind auch nützlich, um schnell ungewichtete Kreuztabellen zu erstellen."

Wie du vielleicht vermutet hast, haben wir Funktionalität zum Erstellen von Pivot-Tabellen für unsere Verwendung in pandas verfügbar. Die Art und Weise, wie wir dies tun, ist, indem wir die `pivot_table()`-Funktion aufrufen, die auf dem pandas-Modul verfügbar ist (das wir als `pd` gespeichert haben). Wie uns die [Dokumentation](https://pandas.pydata.org/pandas-docs/stable/user_guide/reshaping.html) sagt, erwartet `pivot_table()` eine Reihe verschiedener Argumente:

1. `data`: Ein DataFrame-Objekt
2. `values`: eine Spalte oder eine Liste von Spalten zum Aggregieren
3. `index`: eine Spalte, Grouper, Array, das die gleiche Länge wie die Daten hat, oder Liste davon. Schlüssel zum Gruppieren im Pivot-Tabellenindex. Wenn ein Array übergeben wird, wird es auf die gleiche Weise wie Spaltenwerte verwendet.
4. `columns`: eine Spalte, Grouper, Array, das die gleiche Länge wie die Daten hat, oder Liste davon. Schlüssel zum Gruppieren in der Pivot-Tabellenspalte. Wenn ein Array übergeben wird, wird es auf die gleiche Weise wie Spaltenwerte verwendet.
5. `aggfunc`: Funktion zur Verwendung für Aggregation, standardmäßig np.mean

Beachte, dass dies standardmäßig den Mittelwert für den `aggfunc`-Parameter verwendet.

```python
# Let's recall what the data looks like. 
import pandas as pd
red_wines_df = pd.read_csv('data/winequality-red.csv', sep=';')
red_wines_df.head()
```

Lass uns einen Moment nehmen, um schnell eine andere Pandas-Funktion namens `cut()` kennenzulernen, die es uns ermöglicht, eine Spalte mit kontinuierlichen Daten in kategorische Daten umzuwandeln, indem wir Bins angeben, in die sie platziert werden sollen.

```python
pd.cut(red_wines_df['fixed acidity'], bins=np.arange(4, 17)).head()
```

```python
fixed_acidity_bins = np.arange(4, 17)
fixed_acidity_series = pd.cut(red_wines_df['fixed acidity'], bins=fixed_acidity_bins, 
                              labels=fixed_acidity_bins[:-1])
fixed_acidity_series.name = 'fa_bin'
red_wines_df = pd.concat([red_wines_df, fixed_acidity_series], axis=1)
```

```python
red_wines_df.head()
```

Jetzt können wir den mittleren Restzucker für jede Qualitätskategorie/fixed acidity bin erhalten, wie wir es zuvor getan haben, aber mit einer pivot_table (Mittelwert ist die standardmäßige Aggregationsfunktion).

```python
pd.pivot_table(red_wines_df, values='residual sugar', index='quality', columns='fa_bin')
```

```python
# We can also specify a function to aggregate with (by default it is mean)
pd.pivot_table(red_wines_df, values='residual sugar', index='quality', 
               columns='fa_bin', aggfunc=np.max)
```
