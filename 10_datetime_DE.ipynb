### Python Datetime-Modul

Zuerst werden wir uns das datetime-Modul von Python ansehen. Beginnen wir einfach. Wir werden uns die Grundlagen des Moduls ansehen und wie man Daten und Zeit definiert.

```python
# We create a date variable for a random date with the datetime module and as a string…
d1 = date(2021, 5, 14)
d2 = '2021-05-14'

# …and print out the variables. Do you see any differences?
print(d1)
print(d2)
```

```python
# If we have a look at the types of those two variables we can see that there is a difference.
print(type(d1))
print(type(d2))
```

```python
# You can also easily create a variable with the date of today.
d3 = date.today()
print('Date of today: ',d3)

# You can print the day…
print('Day: ', d3.day)

# …or month…
print('Month: ', d3.month)

# …or year.
print('Year: ', d3.year)
```

```python
# The same applies to the time method.
t1 = time(13, 20, 13, 40)
t2 = '13:20:13.000040'
print(t1)
print(t2)
print(type(t1))
print(type(t2))
```

```python
# You can print the hour…
print('Hour :',t1.hour)

# …or minute…
print('Minute :',t1.minute)

# …or second…
print('Second :',t1.second)

# …and even the microseconds.
print('Microsecond :',t1.microsecond)
```

Oben haben wir uns das Datum und die Zeit separat mit den Methoden `date` und `time` angesehen. Aber wir können auch beides mit der `datetime`-Methode kombinieren.

```python
# We create a variable with a random date and time with the datetime method…
dt1 = datetime(2021, 5, 14, 11, 20, 30, 40)

# …and print it.
print(dt1)
print(type(dt1))
```

```python
# You can also create a variable with your local date-time.
dt2 = datetime.now()
dt2
```

Eine wirklich coole Sache, die du mit der DateTime-Funktion machen kannst, ist, den Wochentag zu extrahieren! Dies ist besonders hilfreich beim Feature Engineering, weil der Wert der Zielvariablen vom Wochentag abhängig sein kann, wie z.B. Verkäufe eines Produkts sind im Allgemeinen am Wochenende höher oder der Verkehr auf StackOverflow könnte an einem Wochentag höher sein, wenn Menschen arbeiten, usw.

>Hinweis: `.weekday` gibt den Wochentag zurück, beginnend mit 0 für Montag und endend mit 6 für Sonntag. `.isoweekday` hingegen gibt den Wochentag zurück, wo Montag 1 und Sonntag 7 ist.

```python
# We can define the date of today…
dt2 = datetime.now()

# …and print the weekday where Monday is 0…
print(dt2.weekday()) # outputs e.g. 3 for Thursday

# …or where Monday is 1.
print(dt2.isoweekday()) # outputs 4 in ISO format
```

Ein weiteres sehr wichtiges Feature, das du aus dem gegebenen Datum in einem Datensatz generieren kannst, ist die Woche des Jahres.

```python
# We can define the date of today…
dt2 = datetime.now()

# …and return the year, week and weekday. 
print('Date:', dt2.isocalendar())
print('Week :', dt2.isocalendar()[1])
```

#### Strptime

Mit `strptime` und der richtigen Kombination der Formatcodes (siehe Liste oben) kannst du einen String in einen datetime-Typ konvertieren:

```python
# Create a variable with a datetime stored as string…
date = '22 April, 2020 13:20:13'

# …and convert the string to datetime type.
d1 = datetime.strptime(date,'%d %B, %Y %H:%M:%S')
print(d1)
print(type(d1))
```

```python
# We can create a date stored as string…
date = '2021-05-14'
print(date)
print(type(date))

# …and convert the string to datetime…
d2 = datetime.strptime(date, '%Y-%m-%d')
print(d2)
print(type(d2))

# …or convert the string to date format. 
d2 = datetime.strptime(date, '%Y-%m-%d').date()
print(d2)
print(type(d2))
```

#### Strftime

Wenn wir eine `datetime`-Variable in einen String konvertieren möchten, können wir `.strftime` und die richtige Kombination von Formatcodes verwenden:

```python
# Create a datetime variable…
d1 = datetime.now()
print('Datetime object :',d1)

# …and convert it into a string.
new_date = d1.strftime('%d/%m/%Y %H:%M')
print('Formatted date :',new_date)
print(type(new_date))
```

### Rechnen mit dem datetime-Modul

Manchmal musst du möglicherweise die Zeitspanne zwischen zwei Daten finden, was ein weiteres sehr nützliches Feature sein kann, das du aus einem Datensatz ableiten kannst. Durch Subtraktion eines Datums von einem anderen erhältst du eine `timedelta`.

```python
# timedelta : duration between dates
d1 = datetime(2020,4,23,11,13,10)
d2 = datetime(2021,4,23,12,13,10)
duration = d2 - d1
print(type(duration))
duration
```

Wie du sehen kannst, wird die Dauer als Anzahl der Tage für das Datum und Sekunden für die Zeit zwischen den Daten zurückgegeben. Für deine Features kannst du diese Werte auch separat abrufen:

```python
print('Duration:', duration)
print('Days:', duration.days) # 365
print('Seconds:' ,duration.seconds) # 3600
```

Aber was ist, wenn du tatsächlich die Dauer in Stunden oder Minuten benötigst? Nun, es gibt eine einfache Lösung dafür.

`timedelta` ist auch eine Klasse im `datetime`-Modul. Du kannst sie also verwenden, um deine Dauer in Stunden und Minuten umzuwandeln, wie der folgende Code zeigt:

```python
# duration in hours
print('Duration in hours :', duration / timedelta(hours = 1))

# duration in minutes
print('Duration in minutes :', duration / timedelta(minutes = 1))

# duration in seconds
print('Duration in seconds :', duration / timedelta(seconds = 1))
```

`timedelta` macht es auch möglich, Integer von einem `datetime`-Objekt zu addieren und zu subtrahieren.

```python
# Create datetime variable for current date…
d1 = datetime.now().date()
print("Today's date :",d1)

# …and add two days…
d2 = d1+timedelta(days=2)
print("Date 2 days from today :",d2)

# …or add two weeks.
d3 = d1+timedelta(weeks=2)
print("Date 2 weeks from today :",d3)
```

## DateTime in pandas

Wir wissen bereits, dass pandas eine großartige Bibliothek für Datenanalyseaufgaben ist. Und so versteht es sich von selbst, dass pandas auch Python DateTime-Objekte unterstützt. Es hat einige großartige Methoden für den Umgang mit Daten und Zeiten, wie `to_datetime()` und `to_timedelta()`.

```python
# Define a date using .to_datetime()
date = pd.to_datetime('24th of April, 2020')
print(date)
print(type(date))
```

Du hast vielleicht etwas Seltsames bemerkt. Der Typ des von `.to_datetime()` zurückgegebenen Objekts ist nicht DateTime, sondern Timestamp. Nun, keine Sorge, es ist nur das pandas-Äquivalent zu Pythons DateTime.

Wir wissen bereits, dass time delta Unterschiede in Zeiten angibt. Die Pandas `.to_timedelta()`-Methode tut dasselbe:

```python
# Define and print a variable with the current date and time…
date = datetime.now()
print(date)

# …and print the date of tomorrow…
print(date + pd.to_timedelta(1, unit = 'D'))

# …and next week.
print(date + pd.to_timedelta(1, unit = 'W'))
```

### Date Range in pandas

Um die Erstellung von Datumsfolgen zu einer bequemen Aufgabe zu machen, bietet pandas die `.date_range()`-Methode. Sie akzeptiert ein Startdatum, ein Enddatum und einen optionalen Häufigkeitscode:

```python
# Create a range of dates for 1 month
pd.date_range(start='24/4/2020', end='24/5/2020', freq='D')
```

Anstatt das Enddatum zu definieren, könntest du auch die Periode oder Anzahl der Zeitperioden definieren, die du generieren möchtest: [Siehe hier für eine Liste der Häufigkeitsaliase.](https://pandas.pydata.org/docs/user_guide/timeseries.html#timeseries-offset-aliases)

```python
# Define variable with date of today…
start_date = datetime.today()

# …and create sequence with consecutive minutes…  
dates_start = pd.date_range(start=start_date, periods=10, freq='T')
dates_start
```

```python
# …or consecutive days. 
dates_end = pd.date_range(start=start_date, periods=10, freq='D')
dates_end
```

Lass uns einen Blick darauf werfen, wie man neue Features aus `datetime/timestamp`-Spalten in einem pandas DataFrame erstellt. Daher werden wir einen kleinen DataFrame erstellen, der das `start_date` und `end_date` (siehe Zellen oben) als Features und eine Zielspalte mit drei verschiedenen Klassen enthält.

```python
# Create target column with random classes (1, 2 or 3)
import random
randomList = []
for i in range(10):
    randomList.append(random.randint(1,3))

# Create DataFrame out of previously defined variables
df = pd.DataFrame()
df['Start_date'] = dates_start
df['End_date'] = dates_end
df['Target'] = randomList

# Show first 5 rows 
df.head()
```

```python
# The elements in the date columns are still timestamps 
print(df.Start_date[0])
print(type(df.Start_date[0]))
```

Wir können mehrere neue Features aus der Datumsspalte erstellen, wie Tag, Monat, Jahr, Stunde, Minute usw., indem wir das `dt`-Attribut verwenden. Lass uns zum Beispiel eine neue Spalte erstellen, die die Information über den Tag unserer `end_date`-Spalte enthält.

```python
# Create new column with day of end_date column
df['Day_of_end_date'] = df['End_date'].dt.day
df.head(10)
```

```python
# Create new column for year of start_date column
df['Year_of_start_date'] = df['Start_date'].dt.year
df.head()
```
