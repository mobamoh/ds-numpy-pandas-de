```python
# import pandas
import pandas as pd
from IPython.display import display
```

```python
# load the data
df = pd.read_csv('data/iris.csv')
```

Lass uns zunächst einen Blick auf den Anfang der Tabelle werfen, vielleicht auch auf die letzten 10 Zeilen...

```python
display(df.head())
display(df.tail(10))
```

Wie viele Iris-Pflanzen sind im Datensatz?

```python
len(df)
```

Wie viele verschiedene Arten gibt es?

```python
df["species"].nunique()
```

Berechne Mittelwert, Median, Modus für die Blütenblattlänge

```python
petal_length = df["petal_length"]
pd.Series({
    "mean": petal_length.mean(),
    "median": petal_length.median(),
    "mode": petal_length.mode().iloc[0],
})
```

Was kannst du über die Verteilung der Blütenblattlänge schlussfolgern?

```python
print("Die Verteilung der Blütenblattlänge wirkt nicht symmetrisch, sondern eher mehrgipflig.")
print("Das passt dazu, dass die drei Iris-Arten unterschiedlich große Blütenblätter haben.")
```

Was ist der kleinste und größte Wert für die Blütenblattlänge?

```python
petal_length.min(), petal_length.max()
```

Berechne die Varianz und Standardabweichung für die Blütenblattlänge

```python
pd.Series({
    "variance": petal_length.var(),
    "standard_deviation": petal_length.std(),
})
```

Berechne die grundlegenden deskriptiven Statistiken für alle Spalten des Iris-Datensatzes mit einem einzigen Befehl.

```python
df.describe(include="all")
```

Was ist der Gesamtdurchschnitt für die Kelchblattlänge?

```python
df["sepal_length"].mean()
```

Verwende die DataFrame-Gruppierungsfunktion, um die Anzahl nach Arten zu bestimmen.

```python
df.groupby("species").size()
```

Verwende die DataFrame-Gruppierungsfunktion, um die durchschnittliche Länge und Breite von Kelchblättern und Blütenblättern nach Arten zu bestimmen.

```python
df.groupby("species")[["sepal_length", "sepal_width", "petal_length", "petal_width"]].mean()
```

Füge die Summe der Kelchblattbreite und -länge als neue Spalte zu deinem DataFrame hinzu.

```python
df["sepal_total"] = df["sepal_length"] + df["sepal_width"]
df.head()
```

Visualisiere die Blütenblattlänge mit einem Histogramm

```python
df["petal_length"].plot(kind="hist", bins=20, title="Histogramm der Blütenblattlänge")
```

Beschreibe die Verteilung

```python
print("Das Histogramm zeigt eine mehrgipflige Verteilung mit klaren Gruppen statt einer einzelnen Glockenkurve.")
print("Die Blütenblattlängen der Arten überlappen nur teilweise.")
```

Was ist informativer: die zusammenfassenden Statistiken oder Visualisierungen?

```python
print("Zusammenfassende Statistiken sind gut für exakte Kennzahlen wie Mittelwert oder Streuung.")
print("Visualisierungen sind hier informativer, weil sie Form, Gruppen und mögliche Cluster direkt zeigen.")
```

Visualisiere Blütenblattlänge und -breite mit einer "scatter_matrix"

```python
pd.plotting.scatter_matrix(df[["petal_length", "petal_width"]], figsize=(8, 8), diagonal="hist")
```

Erstelle eine neue Spalte mit einer groben Schätzung der Blütenblattfläche, indem du Blütenblattlänge und -breite miteinander multiplizierst.

```python
df["petal_area"] = df["petal_length"] * df["petal_width"]
df.head()
```

Erstelle einen neuen DataFrame mit Blütenblattflächen größer als $1cm^2$.

```python
df_large_petal_area = df[df["petal_area"] > 1]
display(df_large_petal_area.head())
df_large_petal_area.shape
```

Erstelle 3 neue DataFrames, einen für jede Art. Verwende den gesamten Datensatz.

```python
df_setosa = df[df["species"] == "Iris-setosa"]
df_versicolor = df[df["species"] == "Iris-versicolor"]
df_virginica = df[df["species"] == "Iris-virginica"]

pd.Series({
    "setosa_rows": len(df_setosa),
    "versicolor_rows": len(df_versicolor),
    "virginica_rows": len(df_virginica),
})
```
