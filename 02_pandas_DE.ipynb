    * Daten in den DataFrame einlesen

Lass uns jetzt etwas Übung mit dem Zugriff auf bestimmte Teile der Daten bekommen. Wenn du etwas zusätzliche Übung möchtest, versuche, jede der Fragen auf mehr als eine Weise zu beantworten (denn denk daran, wir können unsere Daten oft auf ein paar verschiedene Arten greifen).
