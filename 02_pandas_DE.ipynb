4. Greife alle Zeilen, bei denen der `chlorides`-Wert kleiner als 0,10 ist.

5. Greife nun alle Zeilen, bei denen der `chlorides`-Wert größer als der Mittelwert der Spalte ist (versuche **nicht**, einen hartcodierten Wert für den Mittelwert zu verwenden, sondern stattdessen eine Methode).

```python
ph_filtered[ph_filtered["residual sugar"] < 2.0]
```

### Teil 3 - Weitere Übung

Gehen wir zu etwas komplizierteren Dingen über. Verwende dein Wissen über `groupby`s und `Sortieren`, um Folgendes zu beantworten.

1. Erhalte die durchschnittliche Menge an `chlorides` für jeden `quality`-Wert.

```python
df.groupby("quality")["chlorides"].mean()
```

2. Finde für Beobachtungen mit einem `pH` größer als 3,0 und kleiner als 4,0 den durchschnittlichen `alcohol`-Wert nach `pH`.

```python
df[(df["pH"] > 3.0) & (df["pH"] < 4.0)].groupby("pH")["alcohol"].mean()
```

3. Finde für Beobachtungen mit einem `alcohol`-Wert zwischen 9,25 und 9,5 die höchste Menge an `residual sugar`.

```python
df[df["alcohol"].between(9.25, 9.5)]["residual sugar"].max()
```

4. Erstelle eine neue Spalte namens `total_acidity`, die die Summe von `fixed acidity` und `volatile acidity` ist.

```python
df["total_acidity"] = df["fixed acidity"] + df["volatile acidity"]
df.head()
```

5. Finde die durchschnittliche `total_acidity` für jeden der `quality`-Werte.

```python
df.groupby("quality")["total_acidity"].mean()
```

6. Finde die Top 5 `density`-Werte.

```python
df["density"].nlargest(5)
```

7. Finde die 10 niedrigsten `sulphates`-Werte.

```python
df["sulphates"].nsmallest(10)
```
